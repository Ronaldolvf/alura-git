```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import datetime as dt
```

```python
dados = pd.read_csv('IBOVDiA3.csv')               # Importando os dados dp aquivo "IBOVDiA3.csv".
display(dados)
```

```python
tickers = dados["CD"]                            # Selecinando apenas os tickers das ações.
data_inicial = dt.datetime(2022,4,8)
data_final    = dt.datetime(2022,4,11)
lista = tickers 
df = wb.DataReader (lista, "yahoo", data_inicial, data_final) 
df
```

```python
df['Close'] = df['Close'].pct_change()           # Calculando a variação diária da ação.

test = pd.DataFrame(df['Close'].round(4)*100)    # Selecionando a coluna "Close" e multiplicando por 100 as cotações da linha 01. 
test
```

```python
ibov = test.iloc[1].round(2)                   # Buscando os dados da linha 01.
ibov
```

```python
ibov.to_excel('ibov_cotação.xlsx')       # Exportando via arquivo xlsx.
```
